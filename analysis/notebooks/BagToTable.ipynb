```python
import pandas as pd
import bagpy
from bagpy import bagreader

from os import listdir
from os.path import isfile, join

data_dir = 'venue_1'

#list of all bag files in data_dir
bag_files = [f for f in listdir('../../data/' + data_dir) if isfile(join('../../data/' + data_dir, f))]


for bag in bag_files:
    b = bagreader('../../data/'+ data_dir + '/'+ bag)
    
    #make a csv file for every topic in the bag
    csvfiles = []
    for t in b.topics:
        data = b.message_by_topic(t)
        csvfiles.append(data)
        
    completeDf = pd.DataFrame()

    #creating dataframe for each topic csv 
    for index,n in enumerate(csvfiles):
        topic_name = str(n.split('/')[-1].split('.')[0])
        completeDf = completeDf.append(pd.DataFrame([[topic_name], []], columns=["Topic"]))
        dataframe = pd.read_csv(csvfiles[index])
        completeDf = completeDf.append(dataframe)


    #creating complete csv file
    csv_name = '../../analysis/processed_data'+ b.filename[:-4] +'.csv'
    completeDf.to_csv(csv_name, index=False)
    print(csv_name)
    print(completeDf)
    
```
